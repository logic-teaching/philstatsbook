```python
def bootstrap_se(data, func, reps):
	theta_hat_stars = []
	for i in range(reps):
		resample = choices(data, k=len(data))
		theta_hat_star = func(resample)
		theta_hat_stars.append(theta_hat_star)
	return stdev(theta_hat_stars)
```

```python
print(bootstrap_se(data, mean, 50000))
```

```python
print(bootstrap_se(data, mean, 50000))
```

```python
import pandas as pd

# Load a CSV file
df = pd.read_csv('data/NCBirths2004.csv')

# Now 'df' is a DataFrame containing the data from the file
```

```python
print(df)
```

```python
# Using bracket notation
weights = df.Weight

print(weights)
```

```python
# bootstrap 

theta_hat_stars = []

for i in range(10000):
	resample = choices(weights, k=1009)
	theta_hat_star = mean(resample)
	theta_hat_stars.append(theta_hat_star)
```
