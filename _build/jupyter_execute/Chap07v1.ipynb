```python
from scipy.stats import hypergeom
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for the distribution
M = 20  # Total number of objects
n = 7   # Total number of Type I objects
N = 12  # Number of objects drawn

# Calculate PMF (Probability Mass Function)
rv = hypergeom(M, n, N)
x = np.arange(0, n+1)
pmf_values = rv.pmf(x)

# Plot the PMF
plt.stem(x, pmf_values)
plt.xlabel('x')
plt.ylabel('PMF')
plt.title('PMF of Hypergeometric Distribution')
plt.show()
```

```python
# Define parameters for the distribution

N1 = 8  # Total number of objects
K1 = 4   # Total number of Type I objects
n1 = 4  # Number of objects drawn


N2 = 6  # Total number of objects
K2 = 3   # Total number of Type I objects
n2 = 3  # Number of objects drawn

N3 = 12  # Total number of objects
K3 = 6   # Total number of Type I objects
n3 = 3  # Number of objects drawn
```

```python
from scipy.stats import hypergeom
import numpy as np
import matplotlib.pyplot as plt

v = np.arange(0, max(K1, K2, K3)+1) # values for x-axis
Xpdf = hypergeom(N1, K1, n1).pmf(v) # pdf for X
Xcdf = hypergeom(N1, K1, n1).cdf(v) # cdf for X
Xccdf = 1-Xcdf               # ccdf for X

Ypdf = hypergeom(N2, K2, n2).pmf(v) # pdf for Y
Ycdf = hypergeom(N2, K2, n2).cdf(v) # cdf for Y
Yccdf = 1-Ycdf               # ccdf for Y

Zpdf = hypergeom(N3, K3, n3).pmf(v) # pdf for Z
Zcdf = hypergeom(N3, K3, n3).cdf(v) # cdf for Z
Zccdf = 1-Zcdf               # ccdf for Z

threshold = np.full(len(v), 0.05)

fig, axs = plt.subplots(1, 3, figsize=(10, 5))  # 1 row, 3 columns
fig.suptitle('X~Hypergeom(%i, %i, %i), Y~Hypergeom(%i, %i, %i), Z~Hypergeom(%i, %i, %i)' % (N1, K1, n1, N2, K2, n2, N3, K3, n3), fontsize=10)

 # Plot the PDF on the first subplot
axs[0].set_xticks(v)
axs[0].plot(v, Xpdf, label='X')
axs[0].plot(v, Ypdf, label='Y')
axs[0].plot(v, Zpdf, label='Z')
axs[0].plot(v, threshold, label='5%')
axs[0].set_ylim(0, 1)
axs[0].legend()
axs[0].set_title('pdf')

# Plot the CDF on the second subplot
axs[1].set_xticks(v)
axs[1].plot(v, Xcdf)
axs[1].plot(v, Ycdf)
axs[1].plot(v, Zcdf)
axs[1].set_ylim(0, 1.1)
axs[1].set_title('cdf')

# Plot the CCDF on the third subplot
axs[2].set_xticks(v)
axs[2].plot(v, Xccdf)
axs[2].plot(v, Yccdf)
axs[2].plot(v, Zccdf)
axs[2].set_ylim(0, 1.1)
axs[2].set_title('ccdf')

plt.show()
```
