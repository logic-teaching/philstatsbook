## Mayo on the classical view

Among philosophers, Mayo has argued for a place for classical hypothesis testing in the philosophy of science and in theory choice more broadly. Her view is that:

> Evidence $e$ should be taken as good grounds for $H_1$ to the extent that $H_1$ has passed a *severe* test with $e$ ({cite}`Mayo1996-re` p. 177; changed $H$ to $H_1$ to better match our discussion here).

Let us follow Mayo in writing the evidence as $e$, keeping in mind that in hypothesis testing it will usually have the form of observing $T_n(\omega)\geq c$, where $\omega$ is the actual world. Further, while we have put an $n$-subscript on the test to handle the examples, let us follow Mayo in dropping the $n$-subscript and just write $T$ for the test.

### Severity and low Type I error

What does it mean for $H_1$ to have passed a severe test $T$ by means of evidence $e$? Mayo tells us a few pages later:

$H_1$ passes a severe test $T$ by means of evidence $e$ iff it passes and moreover 

> there is a very low probability that test procedure $T$ would yield such a passing result [for $H_1$], if $H_1$ were false ({cite}`Mayo1996-re` p. 180, Severity criterion 1b; changed $H$ to $H_1$ to better match our discussion here).

To see the connection to our previous discussion:

- As in our previous discussion, suppose that $H_0$ and $H_1$ are genuine contraries, in that one and only one can be true.

- Suppose that test $T$ passes $H_1$ iff test $T$ rejects $H_0$.

With these minor emendations, Mayo's previous quotation reads as follows:

$H_1$ passes a severe test $T$ by means of evidence $e$ iff $H_0$ is rejected by $T$ with evidence $e$ and moreover 

> there is a very low probability that test procedure $T$ would reject $H_0$ if $H_0$ were true.

One way to convince yourself that this reading is right is to look at her statistical example on pp. 192-193. This is just our Example 1 from last Chapter with $n=100$ and $c=60$. She writes $H^{\prime}$ for our $H_1$, and she like us writes $H_0$ for $H_0$.

Hence, Mayo's severe test has as one component just a low probability of a Type I error. The other component is that the evidence actually pass the test.

### The chief virtue of Mayo's account

The chief virtue of Mayo's account is that it tries to identify a more general constraint on the confirmation of hypotheses that is satisfied by the tradition of classical statistics. That is, it tries to identify an overriding rationale for this tradition which makes it fit in a broader and *primae facie* compelling account of theory choice, to wit: only choose hypotheses which have undergone and passed a severe test, one which if the hypothesis were true would have been unlikely to reject it.

### Mayo on underdetermination

Mayo applies this to the problem of the traditional problem of the underdetermination of hypotheses by evidence. Her perferred formulation of the challenge is the plausibility of the following:

> *Methodological underdetermination [MUD]*: any evidence taken as a good test of (or good support for) hypothesis $[H_0]$ would (on that account of testing or support) be taken as an equally good test of (or equally good support for) some rival [$H_1$] to $H_0$ ({cite}`Mayo1996-re` p. 176, changed her $H_1$ and $H_2$ to $H_0$ and $H_1$).


Other traditional views about hypotheses testing have a hard time responding to this challenge:

- Deductive approaches view $H_0$ and its rivals $H_1$ as formal theories that are confirmed by evidence $e$ which is formally entailed by the theories, e.g. $H_i\vdash e$. But it is pretty easy to cook up lots of alternative ways to deduce a given result by e.g. adding on extraneous axioms.

- Bayesian approaches view $H_0$ and its rivials $H_1$ as confirming evidence $e$ by making $P(H_i\mid e)-P(H_i)$ a large number. But it is easy to make this happen for a large number of theories, as long as one is willing to make the sample space big enough so that it has lots of "pockets" of non-zero probability.


Mayo thinks that her approach has an advantage, since on the classical statistical picture the null hypothesis $H_0$ and $H_1$ partition the options. If one is asking about another rival $H_2$ to $H_1$, one is really asking the "wrong question," as far as the test at issue is concerned ({cite}`Mayo1996-re` p. 189). Moreover, what question is right and what is wrong is dictacted by its place within the hierarchy of experimental inquiry. For, experimental inquiry can be

> viewed as a series of models, each with different questions, stretching from low-level theories of data and experiment to higher level hypotheses and theories of interest ({cite}`Mayo1996-re` p. 189, cf. p. 213)


### Mayo and frequentism

But Mayo is a frequentist ({cite}`Mayo1996-re` pp. 181, cf. p. 121) for whom probabilities just report long-run frequencies of outcomes in many repeated experiments. She then is requiring that 

$H_1$ passes a severe test $T$ by means of evidence $e$ iff $H_0$ is rejected by $T$ with evidence $e$ and moreover 

> if $H_0$ were true, then in the long run there is a low relative frequency that $H_0$ is rejected by test $T$.

This may seem natural, since it just requires an external success notion on choice of hypotheses. However, if the external success notion is hard for us to check and be in a position to ascertain, one might worry that it is not a good fit for the epistemic notions of evidence and theory choice which Mayo seems to want to be delivering. But this is a concern not specific to her but is a concern about the broader classical tradition.

## Howson-Urbach's critique of the classical view

Howson-Urbach's main criticism of the classical view is that they offer nothing by way of how their hypothesis tests relate to strength of evidence or inductive support, or to any other clear epistemic considerations or considerations familiar from theory choice in philosophy of science or related fields.

But they also have some more local criticisms.

### On the Neyman-Pearson view

We are again considering the following kind of test:

Test: reject $H_0$ if $T_n\geq c$. 

Suppose that the probabilities of the Type I and Type II errors are low, and suppose that we observe $T_n(\omega)\geq c$ in the actual world $\omega$. 

According to Neyman-Pearson, if we act according to the test, then 

> we insure that, in the long run of experience, we shall not be too often wrong ({cite}`Neyman1933-zd` p. 291)

In other words:

> if we behave according to such a rule [the test], then in the long run we shall reject $H_0$ when it is true not more, say than once in a hundred times   ({cite}`Neyman1933-zd` p. 291; notation $H$ changed to $H_0$ to match preceding.)


Here is the Howson-Urbach critique:

- If the "long run" is intended to be an appeal to frequentism, then the quotations are just recapitulations of the claims the probability of Type I error is low. It is then an "unhelpful truism."

- If the "long run" is not intended as an appeal to frequentism but just as shorthand for "a great many of times," then there's no reason to think that the quotation is justified by the probability of Type I error being low.

(This is my rendering of their argument on {cite}`Howson2006-oy` p. 151, paragraph beginning "This is a surprising ...").


### On the choice of the null hypothesis

Consider the following variation on Example 1 from the last Chapter. 

We are coin flipping again and our two hypotheses are 

$H_0$: $\hspace{5mm}$   $p=.4$

$H_1$:  $\hspace{5mm}$  $p=.6$

We consider the following test:

Test: reject $H_0$ if $T_{100}\geq 50$.

Glancing at the below graphs we see that the probability of the Type I and Type II errors is low there. 

Supposing that we observe  $T_{100}(\omega)\geq 50$ in the actual world $\omega$, we then reject $H_0$.

```python
import numpy as np                # load numpy package
import matplotlib.pyplot as plt   # load pyplot package
from scipy.stats import binom     # load binomial distribution
```

```python

def Binomvisual(n1,p,q):

    v = np.arange(0, n1) # values for x-axis
    Xpdf = binom.pmf(v, n1, p) # pdf for X
    Xcdf = binom.cdf(v, n1, p) # cdf for X
    Xccdf = 1-Xcdf               # ccdf for X
    Ypdf = binom.pmf(v, n1, q) # pdf for Y
    Ycdf = binom.cdf(v, n1, q) # cdf for Y
    Yccdf = 1-Ycdf               # ccdf for Y


    fig, axs = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

    axs[0].plot(v, Xpdf, label='%1.2f null' % p)
    axs[0].plot(v, Ypdf, label='%1.2f alternative' % q)
    axs[0].set_ylim(0, .2)
    axs[0].set_title('pdf of T_%i~ \n Binom(%i, %1.2f) vs. Binom(%i, %1.2f)' % (n1,n1,p, n1, q))
    axs[0].legend()

    axs[1].plot(v, Xpdf+Xccdf, label='Type I')
    axs[1].plot(v, Ycdf, color ='C1', label='Type II')
    axs[1].set_ylim(0, 1.1)
    axs[1].set_title('probability of \n Type I vs. II error \n as function of c')
    axs[1].set_xlabel('c')  # Add x-axis label
    axs[1].axhline(y=.05, color='r')
    axs[1].legend()

```

```python
Binomvisual(100, .4,.6)
```

Howson-Urbach write, speaking of this example vivified not as counting the heads of coin flips but as counting the reds in the growing of red and white tulips:

> if an experiment showed 50 red-flowering plants in a random sample of 100, then [$H_0$] ([$p=.4]$ red) would be rejected at the .05 level if it were the null hypothesis, and [$H_1$] ([$p=.6$] red) would be accepted. But with [$H_1$] as the null hypothesis, the opposite judgement would be delivered! ({cite}`Howson2006-oy` p. 156, their $h_1$ and $h_2$ switched to our $H_0$ and $H_1$, and with the percents changed to e.g. $p=.4$ and $p=.6$).

I take them to be saying that if $H_1$ was the null hypothesis and we made the same observation, then we would reject $H_1$. This is a problem because in this case the two hypotheses are symmetric ("indistinguishable") and thus the decision to reject or accept comes down to an "arbitrary" decision as to which to designate as the null hypothesis.

However, it appears that it is not true that we would reject $p=.6$ were it the null hypothesis and were we to observe 50 heads (or reds). For by glancing down at the diagrams one sees that the probability of both the Type I and Type II error is quite high in these settings. 

```python
Binomvisual(100, .6,.4)
```

### On sample size

Howson-Urbach draw attention to what happens when $n$ becomes large. 

Below we display $n=1000$

Suppose that for $n=1000$ we observed $T_n(\omega)= 427$ and hence $T_n(\omega)\geq 427$ in the actual world $\omega$.

According to the classical view, we should proceed in rejecting the null hypothesis $p=.4$.

But Howson-Urbach claim that

> an unprejudiced consideration would clearly lead to the conclusion that [for such high values of $n$], the supposedly critical values [like 427] *support* [$H_0$, the null hypothesis] more and more ({cite}`Howson2006-oy` p. 154, their $h_1$ switched to our $H_0$).

They don't expand on this, but my sense is that they are referring to the pdf (the diagram to the left), and are pointing out that an observation of $T_n(\omega)=427$ is so much closer to to the center of the pdf of blue $p=.4$ than the orange $p=.6$. Hence, for this reason

> The conclusion of significance tests oftenly flatly contradict those than an impartial scientist or ordinary observer would draw ({cite}`Howson2006-oy` p. 154).



```python
Binomvisual(1000, .4, .6)
```
