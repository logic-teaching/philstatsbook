```python
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values
x = np.linspace(0, 1, 400)

# Calculate corresponding y values
y = 1 - x

# Create the plot
plt.plot(x, y, color = 'purple')
plt.scatter([1/2], [1/2], color='blue')


# Set the labels for x and y axis
plt.xlabel('X')
plt.ylabel('Y')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x, y values
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
x, y = np.meshgrid(x, y)

# Calculate corresponding z
z = 1 - x - y

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, alpha=0.5, rstride=100, cstride=100, color = 'purple')
ax.scatter([1/3], [1/3], [1/3], color='blue')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values
x = np.linspace(0, 1, 400)

# Calculate corresponding y values
y = 1 - x
kl = -(x*np.log(x)+(1-x)*np.log(1-x))
tv = abs(x-1/2)

# Create the plot
plt.plot(x, y, color = 'purple')
plt.plot(x, kl, color = 'green')
plt.plot(x, tv, color = 'orange')
plt.scatter([1/2], [1/2], color='blue')



# Set the labels for x and y axis
plt.xlabel('X')
plt.ylabel('Y')

# Show the plot
plt.show()
```

```python


import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x, y values
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
x, y = np.meshgrid(x, y)

# Calculate corresponding z

epsilon = .01

# Calculate corresponding z
z = 1 - x - y
kl = -(x*np.log(x+epsilon)+y*np.log(y+epsilon)+(1-x-y)*np.log(1-x-y+epsilon))
tv = (1/2)*abs(x-1/2)+(1/2)*abs(y-1/2)+(1/2)*abs(1-x-y-1/2)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, alpha=0.5, rstride=100, cstride=100, color = 'purple')
ax.plot_surface(x, y, kl, alpha=0.5, rstride=100, cstride=100, color = 'green')
ax.plot_surface(x, y, tv, alpha=0.5, rstride=100, cstride=100, color = 'orange')

ax.scatter([1/3], [1/3], [1/3], color='blue')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
